```python
import pandas as pd   #to create dataframe
import requests       #to send the request to the URL
from bs4 import BeautifulSoup #to get the content in the form of HTML
import numpy as np  # to count the values (in our case)
```

```python
from datetime import date
today=date.today()
d=today.strftime("%m-%d-%y")
print("date=",d)
```

```python
url='https://www.cnbc.com/2022/06/01/russia-ukraine-live-updates.html'
url
```

```python

response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
soup
```

```python
for link in soup.findAll("h2"):
    print("Headline:{}".format(link.text))
```

```python
for news in soup.findAll('div',{'class':'PageBuilder-containerFluidWidths PageBuilder-pageRow'}):
    print(news.text)
```
